```python
def summarise(df):
    print('=======================================================================================')
    print('||                                    Summary                                        ||')
    print('=======================================================================================')
    print('Rows:\t\t\t\t', df.shape[0], '\tColumns:\t\t\t\t',df.shape[1])
    print('Null Valued Columns:\t\t', df.isnull().nunique().sum())
    print('Null Valued > 80%:\t\t', ((df.isnull().sum().sort_values(ascending = False)*100/df.shape[0]) > 80).sum())
    
```

```python
summarise(house_train)
```

```python
Nulls = (house_train.isnull().sum().sort_values(ascending = False)*100/house_train.shape[0] ).to_frame()
Nulls.columns = ['Percentage']
```
