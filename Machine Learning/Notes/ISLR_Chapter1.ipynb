### Notation

$n$ = Number of __distinct__ data points, or, simply records/ observations <br>
$p$ = Number of variables for making the predictions, or, simply parameters/ features <br>

The matrix is denoted with $\textbf{X}$, where
<center>$\textbf{X} = x_{ij} $ </center> <br>
$i = 1,2,....,n$  <br>
$j = 1,2,....,p$

```r
set.seed(1)
x = rnorm(50)
```

```r
y = x+rnorm(50, mean=50, sd=0.1)
```

```r
cor(x,y)
```

```r
#sqrt(var(x))
sd(x)
```

```r
plot(x,y)
```

```r
x = seq(0,1, length=10)
```

```r
y=x
f = outer (x ,y , function (x , y ) cos ( y ) /(1+ x ^2) )
contour (x ,y , f )
contour (x ,y ,f , nlevels =45 , add = T )
fa =( f - t ( f ) ) /2
contour (x ,y , fa , nlevels =15)
```

```r
image (x ,y , fa )
persp (x ,y , fa )
persp (x ,y , fa , theta =30)
persp (x ,y , fa , theta =30 , phi =20)
persp (x ,y , fa , theta =30 , phi =70)
persp (x ,y , fa , theta =30 , phi =40)
```
