```python
fib_store[1] = 1
fib_store[2] = 1
```

```python
fib_store
```

```python
def fib(n):
    if n in fib_store.keys():
        return fib_store[n]
    else:
        fib_store[n] = fib(n-1)+fib(n-2)
        return fib_store[n]
```

```python
grid_store = dict()

def gridTraveller(m,n):
    if (m == 1) and (n == 1):
        grid_store[(m,n)] = 1
        
    elif (m==0) or (n== 0):
        grid_store[(m,n)] = 0
        
    else:
        if (m,n) in grid_store.keys():
            return grid_store[(m,n)]
        else:
            grid_store[(m,n)] = gridTraveller(m-1,n) + gridTraveller(m, n-1)
            grid_store[(n,m)] = grid_store[(m,n)]
    
    return grid_store[(m,n)]
        
```

```python
grid_store
```

```python
gridTraveller(100,100)
```
